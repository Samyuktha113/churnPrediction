```python
from scipy.stats import chi2_contingency

# Create a contingency table for state vs churn
contingency_state = pd.crosstab(df['state'], df['churn'])
chi2_state, p_state, _, _ = chi2_contingency(contingency_state)

# Create a contingency table for area code vs churn
contingency_area_code = pd.crosstab(df['area_code'], df['churn'])
chi2_area_code, p_area_code, _, _ = chi2_contingency(contingency_area_code)

print(f'Chi-squared test p-value for state: {p_state}')
print(f'Chi-squared test p-value for area code: {p_area_code}')
```

```python
from sklearn.linear_model import Perceptron
from sklearn.metrics import accuracy_score, classification_report
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import MinMaxScaler
from imblearn.over_sampling import SMOTE

# Assuming X and y are your original features and target variable
# Split the dataset
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Apply Min-Max Scaling
scaler = MinMaxScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)

# Apply SMOTE to balance the training set
smote = SMOTE(random_state=12, sampling_strategy=1.0)
X_train, y_train = smote.fit_resample(X_train, y_train)

# Initialize and train the Perceptron model
model = Perceptron(random_state=42)
model.fit(X_train, y_train)

# Predict on the test set
y_pred = model.predict(X_test)

# Evaluate the model
accuracy = accuracy_score(y_test, y_pred)
report = classification_report(y_test, y_pred)

# Print the results
print(f'Accuracy: {accuracy:.4f}')
print('Classification Report:')
print(report)
```
